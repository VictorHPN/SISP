Script para gerar o arquivo train.txt, que deve ter o path de todos as imgs de treino

```python
import os

%cd /content

imagens = []
os.chdir(os.path.join("data", "obj"))
for filename in os.listdir(os.getcwd()):
    if filename.endswith(".jpeg"):
        imagens.append("data/obj/" + filename)
os.chdir("..")

with open("train.txt", "w") as outfile:
    for image in imagens:
        outfile.write(image)
        outfile.write("\n")
    outfile.close()
os.chdir("..")
```

```python
%cd /content
imagens = []
os.chdir(os.path.join("data", "valid"))
for filename in os.listdir(os.getcwd()):
    if filename.endswith(".jpeg"):
        imagens.append("data/valid/" + filename)
os.chdir("..")

with open("valid.txt", "w") as outfile:
    for image in imagens:
        outfile.write(image)
        outfile.write("\n")
    outfile.close()
os.chdir("..")
```

```python
%cd /content
imagens = []
os.chdir(os.path.join("data", "test"))
for filename in os.listdir(os.getcwd()):
    if filename.endswith(".jpeg"):
        imagens.append("data/test/" + filename)
os.chdir("..")

with open("test.txt", "w") as outfile:
    for image in imagens:
        outfile.write(image)
        outfile.write("\n")
    outfile.close()
os.chdir("..")
```
